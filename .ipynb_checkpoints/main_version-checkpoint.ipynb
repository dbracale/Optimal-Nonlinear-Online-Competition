```python
import numpy as np
import matplotlib.pyplot as plt
import time

from scipy.optimize import minimize, NonlinearConstraint, minimize_scalar
from scipy.interpolate import interp1d
from scipy.stats import truncnorm, norm
from sklearn.isotonic import IsotonicRegression  # (only used for reference/comparison)
from joblib import Parallel, delayed  # for parallelization
from tqdm import tqdm  # for progress bars
```

```python
###############################################
# 1. LS ALGORITHM FOR NON-PARAMETRIC ISOTONIC FIT
###############################################

# Tolerance for near-zero transformation parameter
tol_alpha = 1e-7

def d_alpha_fun(alpha):
    """
    Transformation d_alpha.
    For |alpha| < tol_alpha, use log; otherwise use y**alpha (with sign if alpha < 0).
    """
    if abs(alpha) < tol_alpha:
        return lambda y: np.log(y)
    else:
        if alpha < 0:
            return lambda y: -np.power(y, alpha)
        else:
            return lambda y: np.power(y, alpha)

def d_alpha_inv_fun(alpha):
    """
    Inverse transformation.
    For |alpha| < tol_alpha, use exp; otherwise use the inverse power.
    """
    if abs(alpha) < tol_alpha:
        return lambda x: np.exp(x)
    else:
        if alpha < 0:
            return lambda x: np.power(-x, 1/alpha)
        else:
            return lambda x: np.power(x, 1/alpha)

def convex_matrix(w):
    """
    Constructs a matrix A so that for phi = [phi_1,...,phi_n],
    A.dot(phi) yields the slopes:
         A[i, i]   = -1/(w[i+1]-w[i])
         A[i, i+1] =  1/(w[i+1]-w[i])
    A small cutoff is used to avoid division by zero.
    """
    n = len(w)
    A = np.zeros((n-1, n))
    diffs = np.diff(w)
    diffs = np.where(np.abs(diffs) < 1e-10, 1e-10, diffs)
    for i in range(n-1):
        A[i, i] = -1.0 / diffs[i]
        A[i, i+1] = 1.0 / diffs[i]
    return A

def objective_function(phi, y_s, d_alpha_inv):
    """
    Objective function for LS:
      Sum_i [ y_s[i] - d_alpha_inv(phi[i]) ]^2
    """
    transformed = d_alpha_inv(phi)
    return np.sum((y_s - transformed)**2)

def build_constraints(phi, w, d_alpha, epsi):
    """
    Constructs all constraints in the form c(phi) <= 0:
    
      (i) Value constraints: 
          d_alpha(epsi) <= phi <= d_alpha(1-epsi)
          → c_upper: phi - d_alpha(1-epsi) <= 0
          → c_lower: d_alpha(epsi) - phi <= 0

      (ii) Monotonicity: Let s = A.dot(phi); we require s >= 0, i.e.
          -s <= 0

      (iii) Concavity: The differences between slopes must be non-positive:
          diff(s) <= 0
    """
    # Value constraints
    c_upper = phi - d_alpha(1 - epsi)
    c_lower = d_alpha(epsi) - phi
    
    # Monotonicity & concavity constraints
    A = convex_matrix(w)
    slopes = A.dot(phi)
    c_monotonicity = -slopes          # slopes >= 0 → -slopes <= 0
    c_concavity = np.diff(slopes)      # successive differences <= 0

    return np.concatenate([c_upper, c_lower, c_monotonicity, c_concavity])

def LS(w, y, lamb, epsi):
    """
    Risolve il problema di minimi quadrati con vincoli di monotonicità e concavità
    della trasformazione d_alpha.

    Parametri:
      - w: vettore degli input (ordinato)
      - y: vettore dei valori osservati
      - lamb: parametro di trasformazione
      - epsi: parametro per evitare problematiche ai bordi

    Restituisce:
      Una funzione interpolata F_hat.
    """
    # Definizione delle trasformazioni
    d_alpha = d_alpha_fun(lamb)
    d_alpha_inv = d_alpha_inv_fun(lamb)
    
    # Considera solo i valori unici in w
    w_s, idx = np.unique(w, return_index=True)
    y_s = y[idx]
    n = len(w_s)
    
    # Definisci la funzione obiettivo e i vincoli
    obj_func = lambda phi: objective_function(phi, y_s, d_alpha_inv)
    constr_fun = lambda phi: build_constraints(phi, w_s, d_alpha, epsi)
    
    num_constraints = 2 * n + (n - 1) + (n - 2)
    lower_bounds = -np.inf * np.ones(num_constraints)
    upper_bounds = np.zeros(num_constraints)
    
    nonlinear_constraint = NonlinearConstraint(constr_fun, lower_bounds, upper_bounds)
    
    # Punto iniziale: una progressione lineare tra i limiti
    phi_init = np.linspace(d_alpha(epsi), d_alpha(1 - epsi), n)
    
    # Opzioni aumentate: più iterazioni
    options = {'maxiter': 10000, 'xtol': 1e-8}
    
    result = minimize(obj_func, phi_init, method='trust-constr',
                      constraints=[nonlinear_constraint], options=options)
    
    if not result.success:
        print("LS optimization not converged:", result.message)
    
    phi_opt = result.x
    F_hat = interp1d(w_s, d_alpha_inv(phi_opt), kind='linear', fill_value='extrapolate')
    return F_hat

def LS_SLSQP(w, y, lamb, epsi):
    """
    Versione della funzione LS che usa il metodo 'SLSQP' per risolvere il problema
    di minimi quadrati con vincoli di monotonicità e concavità.
    """
    # Definisce le trasformazioni
    d_alpha = d_alpha_fun(lamb)
    d_alpha_inv = d_alpha_inv_fun(lamb)
    
    # Usa solo valori unici in w (e corrispondenti y)
    w_s, idx = np.unique(w, return_index=True)
    y_s = y[idx]
    n = len(w_s)
    
    # Funzione obiettivo
    obj_func = lambda phi: objective_function(phi, y_s, d_alpha_inv)
    
    # Numero totale di vincoli: 2 per il vincolo sui valori, (n-1) per la monotonicità e (n-2) per la concavità.
    num_constraints = 2 * n + (n - 1) + (n - 2)
    
    # Costruisce una lista di vincoli in formato SLSQP:
    constraints_list = [
        {'type': 'ineq', 'fun': lambda phi, idx=i: -build_constraints(phi, w_s, d_alpha, epsi)[idx]}
        for i in range(num_constraints)
    ]
    
    # Punto iniziale: lineare tra i limiti
    phi_init = np.linspace(d_alpha(epsi), d_alpha(1 - epsi), n)
    
    # Opzioni per SLSQP
    options = {'maxiter': 10000, 'ftol': 1e-8}
    
    result = minimize(obj_func, phi_init, method='SLSQP', constraints=constraints_list, options=options)
    
    if not result.success:
        print("LS_SLSQP optimization not converged:", result.message)
    
    phi_opt = result.x
    F_hat = interp1d(w_s, d_alpha_inv(phi_opt), kind='linear', fill_value='extrapolate')
    return F_hat

```

```python
###############################################
# 2. PAYOFF, p_opt AND SUPPORT FUNCTIONS
###############################################

def payoff(psi_func, p, b, q):
    """
    Payoff function: p * psi_func(p * b + q)
    """
    return p * psi_func(p * b + q)

def p_opt(F, b, q, p_min, p_max):
    """
    Given a function F (e.g., psi or its estimate) and parameters,
    find the optimal p in [p_min, p_max] that maximizes the payoff.
    """
    def objective(p):
        return -payoff(F, p, b, q)  # negate for maximization
    result = minimize_scalar(objective, bounds=(p_min, p_max), method='bounded')
    return result.x
```

```python
###############################################
# 3. TRUNCATED GAUSSIAN SAMPLING FUNCTIONS
###############################################

def construct_m_and_s(p_m_arr, p_M_arr):
    """
    Compute center and scale parameters from p boundaries.
    """
    p_m_arr = np.array(p_m_arr, dtype=float)
    p_M_arr = np.array(p_M_arr, dtype=float)
    m = 0.5 * (p_m_arr + p_M_arr)  # center
    sigma = np.sqrt(0.5 * (p_M_arr - p_m_arr))
    return m, sigma

def truncated_multivariate_gaussian(m, s, size=1, random_state=None):
    """
    Draw samples from a truncated multivariate Gaussian where the
    elliptical constraint is: sum((x_i-m_i)^2/s_i^2) <= 1.
    """
    rng = np.random.default_rng(random_state)
    N = len(m)
    out = np.zeros((size, N), dtype=float)
    n_accepted = 0
    while n_accepted < size:
        batch_size = size - n_accepted
        proposals = rng.normal(loc=m, scale=s, size=(batch_size, N))
        centered = (proposals - m) / s
        sq_mahal = np.sum(centered**2, axis=1)
        mask = (sq_mahal <= 1.0)
        accepted = proposals[mask]
        n_good = len(accepted)
        if n_good > 0:
            limit = min(n_good, batch_size)
            out[n_accepted:n_accepted+limit] = accepted[:limit]
            n_accepted += limit
    return out

def sample_truncated_multivariate_times_tau(p_m_arr, p_M_arr, tau, random_state=None):
    """
    Generate tau samples from the truncated multivariate Gaussian.
    """
    m, s = construct_m_and_s(p_m_arr, p_M_arr)
    P = truncated_multivariate_gaussian(m, s, size=tau, random_state=random_state)
    return P
```

```python
###############################################
# 4. EXPERIMENT SETUP: TRUE FUNCTIONS & PARAMETERS
###############################################

N = 6  # number of players/components

# Boundaries for p (for each coordinate)
p_m_arr = [0] * N
p_M_arr = [3] * N

# Define the "true" psi_i functions (for data generation & regret measurement)
psi = []

if N == 8:
    for i in range(N):
        psi.append(lambda w, i=i: norm.cdf(3*w/(i+1)))  # using default argument to fix i
else:
    for i in range(N):
        #Each psi_i is defined as norm.cdf(w - i)
        psi.append(lambda w, i=i: norm.cdf(2*w/(i+1)))  # using default argument to fix i
        #psi.append(lambda w, i=i: 1-np.exp(-w/(i+1)))

# Define the true theta for each component
if N ==2:
    beta_i = 0.75

if N ==4:
    beta_i = 0.92

if N ==6:
    beta_i = 0.96
    
if N ==8:
    beta_i = 0.97

theta_true = []
for i in range(N):
    theta = np.ones(N)
    theta[i] = -beta_i
    # For j != i, set them to sqrt((1-beta_i^2)/(N-1))
    for j in range(N):
        if j != i:
            theta[j] = np.sqrt((1-beta_i**2)/(N-1))
    theta_true.append(theta)


def check_sum_abs(vector):
    sum_abs = np.sum(np.abs(vector))
    threshold = 1 / np.sqrt(2)
    return sum_abs <= threshold

for i in range(N):
    print(check_sum_abs(np.delete(theta_true[i], i)))  # Output: True or False
```

```python
# Compute p* (the optimal p vector) via a fixed-point iteration
p_initial = [p_M_arr[i] / 2 for i in range(N)]
error = 100
while error > 0.001:
    p_t = []
    for i in range(N):
        b_i = theta_true[i][i]
        gamma_i = np.delete(theta_true[i], i)
        p_minus_i = np.delete(p_initial, i)
        q_i = p_minus_i @ gamma_i
        p_t.append(p_opt(psi[i], b_i, q_i, p_m_arr[i], p_M_arr[i]))
    error = np.linalg.norm(np.array(p_initial) - np.array(p_t), 2)
    print("Fixed-point error:", error)
    p_initial = p_t

p_star = p_initial  # optimal p vector
print(p_star)
#p_star = [-1/theta_true[i][i] for n in range(N)]
```

```python
###############################################
# 5. EXPERIMENT: SINGLE RUN EXPERIMENT FUNCTION
###############################################

def single_run_experiment(T, rng_seed):
    """
    Runs a single experiment for T time periods.
    Data is collected, theta and psi are estimated, and regret is computed.
    """
    rng = np.random.default_rng(rng_seed)
    m, s = construct_m_and_s(p_m_arr, p_M_arr)

    # Define tau (number of initial samples) and n_i (for psi estimation)
    tau = int(np.ceil(T**(5/7) / 1.5))
    n_i = int(np.ceil(tau / 2))

    # Prepare data arrays
    P = np.zeros((tau, N))
    Y = np.zeros((tau, N))
    W = np.zeros((tau, N))

    theta_hat = None
    psi_hat = None
    p_T_final = None
    regrets = np.zeros(T)
    l2_error_psi = np.nan  # will be computed later

    # Data collection and updates
    for t in range(T):
        total_payoff_star = 0.0
        total_payoff_chosen = 0.0

        # 1) Data collection for t < tau
        if t < tau:
            # Sample p from the truncated multivariate Gaussian
            p_t_sample = sample_truncated_multivariate_times_tau(p_m_arr, p_M_arr, 1, random_state=rng)[0]
            P[t, :] = p_t_sample
            
            for i in range(N):
                w_t_i = np.dot(theta_true[i], p_t_sample)
                noise_t_i = rng.uniform(-0.03, 0.03)
                y_t_i = psi[i](w_t_i) + noise_t_i
                Y[t, i] = y_t_i
                W[t, i] = w_t_i

                b_i = theta_true[i][i]
                gamma_i = np.delete(theta_true[i], i)
                q_i = np.delete(p_star, i) @ gamma_i
                payoff_star = payoff(psi[i], p_star[i], b_i, q_i)
                payoff_chosen = payoff(psi[i], p_t_sample[i], b_i, q_i)

                total_payoff_star += payoff_star
                total_payoff_chosen += payoff_chosen

            regrets[t] = total_payoff_star - total_payoff_chosen

        # 2) Estimate theta at t == n_i using all data so far
        if t == n_i:
            X_for_theta = P[:n_i,:]
            X_centered = X_for_theta - X_for_theta.mean(axis=0, keepdims=True)
            #P_centered = P - m
            theta_hat_temp = []
            for i in range(N):
                beta_hat_i = np.linalg.lstsq(X_centered, Y[:n_i, i], rcond=None)[0]
                norm_beta = np.linalg.norm(beta_hat_i, 2)
                if norm_beta > 0:
                    beta_hat_i /= norm_beta
                theta_hat_temp.append(beta_hat_i)
            theta_hat = theta_hat_temp

        # 3) Estimate psi at t == tau (using data from rows [n_i, tau))
        if t == tau:
            psi_hat_temp = []
            for i in range(N):
                w_i = W[n_i:tau, i]
                y_i = Y[n_i:tau, i]

                # Instead of using sklearn.IsotonicRegression, use LS (with lambda=0 and epsi=0.001)
                #iso_reg = IsotonicRegression(increasing=True, out_of_bounds='clip')
                #iso_reg.fit(w_i, y_i)
                #iso_reg = LS(w_i, y_i, lamb=0, epsi=0.001)
                iso_reg = LS_SLSQP(w_i, y_i, lamb=0, epsi=0.001)
                
                # Define psi_hat_i that clips the output in [0,1]
                psi_hat_i = lambda u, reg=iso_reg: np.clip(reg(u), 0, 1)
                # psi_hat_i = lambda u, reg=iso_reg: np.clip(reg.predict(np.array([u]))[0], 0, 1)
                psi_hat_temp.append(psi_hat_i)
            psi_hat = psi_hat_temp

            # Compute L2 error of psi estimation (on training data)
            errors = []
            for i in range(N):
                w_i_train = W[n_i:tau, i]
                y_true_i_train = np.array([psi[i](w_val) for w_val in w_i_train])
                y_pred_i_train = np.array([psi_hat[i](w_val) for w_val in w_i_train])
                mse_i = np.mean((y_true_i_train - y_pred_i_train) ** 2)
                errors.append(mse_i)
            l2_error_psi = np.mean(errors)

        # 4) For t >= tau, update p_t using estimated psi (and theta_hat if available)
        if t >= tau:
            if t == tau:
                p_t_minus_1 = p_initial  # starting point for p_t
            p_t_list = []
            for i in range(N):
                p_hat_minus_i = np.delete(p_t_minus_1, i)

                # Use estimated psi if available; otherwise, default to midpoint
                if psi_hat is not None and theta_hat is not None:
                    b_hat_i = theta_hat[i][i]
                    gamma_hat_i = np.delete(theta_hat[i], i)
                    q_hat = p_hat_minus_i @ gamma_hat_i
                    p_hat_t_i = p_opt(psi_hat[i], b_hat_i, q_hat, p_m_arr[i], p_M_arr[i])
                else:
                    p_hat_t_i = (p_m_arr[i] + p_M_arr[i]) / 2.0

                ### regret ###
                b_i = theta_true[i][i]
                gamma_i = np.delete(theta_true[i], i)
                q_i = np.delete(p_star, i) @ gamma_i
                payoff_star = payoff(psi[i], p_star[i], b_i, q_i)
                payoff_chosen = payoff(psi[i], p_hat_t_i, b_i, q_i)
                total_payoff_star += payoff_star
                total_payoff_chosen += payoff_chosen

                p_t_list.append(p_hat_t_i)
            p_t_minus_1 = np.array(p_t_list)

            regrets[t] = total_payoff_star - total_payoff_chosen

            if t == T - 1:
                for i in range(N):
                    b_i = theta_true[i][i]
                    gamma_i = np.delete(theta_true[i], i)
                    b_hat_i = theta_hat[i][i]
                    q_i = np.delete(p_star, i) @ gamma_hat_i
                    gamma_hat_i = np.delete(theta_hat[i], i)
                    q_hat_i = np.delete(p_t_minus_1, i) @ gamma_hat_i
                    p_range = np.linspace(p_m_arr[i], p_M_arr[i], 200)
                    rev_hat  = [payoff(psi_hat[i], p, b_hat_i, q_hat_i) for p in p_range]
                    rev_true = [payoff(psi[i], p, b_i, q_i) for p in p_range]
                
                    # plt.figure(figsize=(8, 5))
                    # plt.plot(p_range, rev_hat, 'r--', lw=2, label='revenue estimate')
                    # plt.plot(p_range, rev_true, 'g-',  lw=2, label='True revenue')
                    # plt.xlabel("p")
                    # plt.ylabel("rev")
                    # plt.legend()
                    # plt.grid(True)
                    # plt.show()
                
                p_T_final = p_t_minus_1

    # After time T, compute performance metrics
    if theta_hat is not None:
        sum_diff_theta = sum(np.linalg.norm(theta_hat[i] - theta_true[i], 2) for i in range(N))
    else:
        sum_diff_theta = np.nan

    if p_T_final is not None and p_star is not None:
        diff_p = np.linalg.norm(p_T_final - p_star, 2)
    else:
        diff_p = np.nan

    total_regret_run = np.nansum(regrets)
    return sum_diff_theta, l2_error_psi, diff_p, total_regret_run
```

```python
###############################################
# 6. RUN EXPERIMENTS IN PARALLEL AND PLOT RESULTS
###############################################

# Define experiment parameters
T_values = [100, 200, 400, 800, 1600]
#T_values = [150, 400, 1600]

num_runs = 35  # number of repetitions per T
rng_master = np.random.default_rng(123)  # master RNG for reproducibility

mean_theta_diffs = []
ci_theta_diffs = []
mean_p_diffs = []
ci_p_diffs = []
mean_regrets = []
ci_regrets = []
mean_l2_errors = []
ci_l2_errors = []

run_parallel = True

for T in T_values:
    print(f"\nRunning experiments for T = {T}")
    seeds = rng_master.integers(1_000_000, size=num_runs)
    if run_parallel == True:
        results = Parallel(n_jobs=-1)(
            delayed(single_run_experiment)(T, seed)
            for seed in tqdm(seeds, desc=f"Repetitions for T={T}", leave=False)
        )
    else:
        results = []
        for seed in seeds:
            results.append(single_run_experiment(T, seed))
    
    run_theta_diffs = np.array([r[0] for r in results])
    run_l2_errors   = np.array([r[1] for r in results])
    run_p_diffs     = np.array([r[2] for r in results])
    run_total_regret= np.array([r[3] for r in results])
    
    def mean_ci(a):
        m = np.nanmean(a)
        s = np.nanstd(a, ddof=1)
        c = 1.96 * s / np.sqrt(num_runs)
        return m, c

    mean_theta, ci_theta = mean_ci(run_theta_diffs)
    mean_p,     ci_p     = mean_ci(run_p_diffs)
    mean_r,     ci_r     = mean_ci(run_total_regret)
    mean_l2,    ci_l2    = mean_ci(run_l2_errors)
    
    mean_theta_diffs.append(mean_theta)
    ci_theta_diffs.append(ci_theta)
    mean_p_diffs.append(mean_p)
    ci_p_diffs.append(ci_p)
    mean_regrets.append(mean_r)
    ci_regrets.append(ci_r)
    mean_l2_errors.append(mean_l2)
    ci_l2_errors.append(ci_l2)
    
    print(f"\n[T={T}] Results over {num_runs} runs:")
    print(f"   Sum(||theta_hat - theta_true||): {mean_theta:.3f} ± {ci_theta:.3f}")
    print(f"   L2 Error (Isotonic via LS): {mean_l2:.5f} ± {ci_l2:.5f}")
    print(f"   ||p_T - p^*||: {mean_p:.3f} ± {ci_p:.3f}")
    print(f"   Total Regret: {mean_r:.3f} ± {ci_r:.3f}")
```

```python
first_nonzero_seen = False  # Flag to track the first nonzero label

def sci_formatter(x, pos):
    global first_nonzero_seen
    if x == 0:
        return "0"
    
    exponent = int(np.floor(np.log10(x)))
    base = x / (10 ** exponent)  # Normalize to 10^exponent
    
    if not first_nonzero_seen:
        first_nonzero_seen = True
        return rf"$10^{{{exponent}}}$"
    else:
        return rf"${int(base)} \times 10^{{{exponent}}}$"  # Force integer base
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

plot_folder = "PLOTS"
os.makedirs(plot_folder, exist_ok=True)

# Use a clean and modern style
plt.style.use('seaborn-v0_8-whitegrid')

mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.size'] = 16
mpl.rcParams['axes.labelsize'] = 18
mpl.rcParams['axes.titlesize'] = 20
mpl.rcParams['legend.fontsize'] = 16
mpl.rcParams['xtick.labelsize'] = 16
mpl.rcParams['ytick.labelsize'] = 14

# Create 4 subplots in a 1x4 layout (wider)
fig_main, axes = plt.subplots(1, 4, figsize=(20, 5))  # Bigger figure

# Flatten the 2D array of axes for easier indexing
axes = axes.ravel()

# (A) L2 Error of theta estimation
ax1 = axes[0]
ax1.errorbar(T_values, mean_theta_diffs, yerr=ci_theta_diffs, fmt='-o', capsize=5, ecolor='red', label='Mean ± 95% CI', linewidth=2, markeredgewidth=1.5, alpha=0.7)
ax1.set_title(r'$\sum_{i=1}^{' + str(N) + r'} \|\widehat{\theta}_i^{(T)} - \theta_i\|_2$', fontsize=20, fontweight='bold')
ax1.set_xlabel('T', fontsize=18)
ax1.set_ylabel(f'N={N}', fontsize=18)
ax1.grid(True, linestyle="--", alpha=0.5)
#ax1.yaxis.set_major_formatter(mpl.ticker.FuncFormatter(sci_formatter))

# (B) MSE of psi estimation
ax2 = axes[1]
ax2.errorbar(T_values, mean_l2_errors, yerr=ci_l2_errors, fmt='-o', capsize=5, ecolor='red', label='Mean ± 95% CI', linewidth=2, markeredgewidth=1.5, alpha=0.7)
ax2.set_title(r'$\sum_{i=1}^{' + str(N) + r'} \|\widehat{\psi}_i^{(T)} - \psi_i\|_2$', fontsize=20, fontweight='bold')
ax2.set_xlabel('T', fontsize=18)
ax2.grid(True, linestyle="--", alpha=0.5)
ax2.yaxis.set_major_formatter(mpl.ticker.FuncFormatter(sci_formatter))

# (C) p vector difference
ax3 = axes[2]
ax3.errorbar(T_values, mean_p_diffs, yerr=ci_p_diffs, fmt='-o', capsize=5, ecolor='red', label='Mean ± 95% CI', linewidth=2, markeredgewidth=1.5, alpha=0.7)
ax3.set_title(r'$\|\mathbf{p}^{(T)} - \mathbf{p}^\star\|_2$', fontsize=20, fontweight='bold')
ax3.set_xlabel('T', fontsize=18)
ax3.grid(True, linestyle="--", alpha=0.5)
#ax3.yaxis.set_major_formatter(mpl.ticker.FuncFormatter(sci_formatter))

# (D) Total Regret
ax4 = axes[3]
ax4.errorbar(T_values, mean_regrets, yerr=ci_regrets, fmt='-o', capsize=5, ecolor='red', label='Mean ± 95% CI', linewidth=2, markeredgewidth=1.5, alpha=0.7)
ax4.set_title("Total Expected Regret", fontsize=20, fontweight='bold')
ax4.set_xlabel('T', fontsize=18)
ax4.grid(True, linestyle="--", alpha=0.5)
#ax4.yaxis.set_major_formatter(mpl.ticker.FuncFormatter(sci_formatter))

plt.tight_layout(pad=3.0)  # More spacing between subplots
plt.subplots_adjust(wspace=0.3, hspace=0.3)  # Adjust spacing between plots

# Save figure
plot_file = os.path.join(plot_folder, f"figure_N_{N}.pdf")
plt.savefig(plot_file, bbox_inches='tight', dpi=300)  # High-res save
plt.show()
print(f"Figure saved successfully to {plot_file}")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# Folder paths
data_folder = "DATA"
plot_folder = "PLOTS"
os.makedirs(data_folder, exist_ok=True)
os.makedirs(plot_folder, exist_ok=True)

# List of N values for which we have data
N_values = [2, 4, 6]

# Set the style and font properties
plt.style.use('seaborn-v0_8-whitegrid')
mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.size'] = 16
mpl.rcParams['axes.labelsize'] = 18
mpl.rcParams['axes.titlesize'] = 20
mpl.rcParams['legend.fontsize'] = 16
mpl.rcParams['xtick.labelsize'] = 16
mpl.rcParams['ytick.labelsize'] = 14

# Optional: scientific notation formatter (if needed)
first_nonzero_seen = False
def sci_formatter(x, pos):
    global first_nonzero_seen
    if x == 0:
        return "0"
    exponent = int(np.floor(np.log10(x)))
    base = x / (10 ** exponent)
    if not first_nonzero_seen:
        first_nonzero_seen = True
        return rf"$10^{{{exponent}}}$"
    else:
        return rf"${int(base)} \times 10^{{{exponent}}}$"

# Create a figure with 3 rows (one for each N) and 4 columns (one for each metric)
fig, axes = plt.subplots(len(N_values), 4, figsize=(20, 12))
axes = axes.reshape(len(N_values), 4)  # ensure axes is 2D

# Loop over each dataset
for i, N in enumerate(N_values):
    # Load the corresponding CSV file
    data_file = os.path.join(data_folder, f"data_N_{N}.csv")
    df = pd.read_csv(data_file)
    
    # Extract the columns from the dataframe
    T_values = df["T_values"]
    mean_theta_diffs = df["mean_theta_diffs"]
    ci_theta_diffs = df["ci_theta_diffs"]
    mean_l2_errors = df["mean_l2_errors"]
    ci_l2_errors = df["ci_l2_errors"]
    mean_p_diffs = df["mean_p_diffs"]
    ci_p_diffs = df["ci_p_diffs"]
    mean_regrets = df["mean_regrets"]
    ci_regrets = df["ci_regrets"]
    
    # (A) Theta estimation error subplot
    ax1 = axes[i, 0]
    ax1.errorbar(T_values, mean_theta_diffs, yerr=ci_theta_diffs, fmt='-o', capsize=5,
                 ecolor='red', linewidth=2, markeredgewidth=1.5, alpha=0.7)
    if N == 2:
        ax1.set_title(r'$\sum_{i=1}^{N} \|\widehat{\theta}_i^{(T)} - \theta_i\|_2$', fontsize=22, fontweight='bold')
    if N == 6:
        ax1.set_xlabel('T', fontsize=20)
    ax1.set_ylabel(f'N={N}', fontsize=30)
    ax1.grid(True, linestyle="--", alpha=0.5)
    # Uncomment if you wish to use the formatter:
    # ax1.yaxis.set_major_formatter(mpl.ticker.FuncFormatter(sci_formatter))
    
    # (B) Psi estimation error subplot
    ax2 = axes[i, 1]
    ax2.errorbar(T_values, mean_l2_errors, yerr=ci_l2_errors, fmt='-o', capsize=5,
                 ecolor='red', linewidth=2, markeredgewidth=1.5, alpha=0.7)
    if N == 2:
        ax2.set_title(r'$\sum_{i=1}^{N} \|\widehat{\psi}_i^{(T)} - \psi_i\|_2$', fontsize=22, fontweight='bold')
    if N == 6:
        ax2.set_xlabel('T', fontsize=20)
    ax2.grid(True, linestyle="--", alpha=0.5)
    ax2.yaxis.set_major_formatter(mpl.ticker.FuncFormatter(sci_formatter))
    
    # (C) p vector difference subplot
    ax3 = axes[i, 2]
    ax3.errorbar(T_values, mean_p_diffs, yerr=ci_p_diffs, fmt='-o', capsize=5,
                 ecolor='red', linewidth=2, markeredgewidth=1.5, alpha=0.7)
    if N == 2:
        ax3.set_title(r'$\|\mathbf{p}^{(T)} - \mathbf{p}^\star\|_2$', fontsize=22, fontweight='bold')
    if N == 6:
        ax3.set_xlabel('T', fontsize=20)
    ax3.grid(True, linestyle="--", alpha=0.5)
    # Uncomment if desired:
    # ax3.yaxis.set_major_formatter(mpl.ticker.FuncFormatter(sci_formatter))
    
    # (D) Total regret subplot
    ax4 = axes[i, 3]
    ax4.errorbar(T_values, mean_regrets, yerr=ci_regrets, fmt='-o', capsize=5,
                 ecolor='red', linewidth=2, markeredgewidth=1.5, alpha=0.7)
    if N == 2:
        ax4.set_title("Total Expected Regret", fontsize=22, fontweight='bold')
    if N == 6:
        ax4.set_xlabel('T', fontsize=20)
    ax4.grid(True, linestyle="--", alpha=0.5)
    # Uncomment if desired:
    # ax4.yaxis.set_major_formatter(mpl.ticker.FuncFormatter(sci_formatter))

plt.tight_layout(pad=3.0)
plt.subplots_adjust(wspace=0.3, hspace=0.2)

# Save and display the combined figure
plot_file = os.path.join(plot_folder, "combined_figure_N_2_4_8.pdf")
plt.savefig(plot_file, bbox_inches='tight', dpi=300)
plt.show()
print(f"Combined figure saved successfully to {plot_file}")
```
